```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
```

```python
# Loading dataset
data, target = load_breast_cancer(return_X_y=True, as_frame=True)
data.head()
```

```python
# Plot the distribution of each feature column in the data
data.hist(figsize=(20, 15))
plt.suptitle("Breast Cancer Dataset - Features Distribution")
plt.show()
```

```python
# Split data into train and test
x, y = data, target
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
# Define and run classifier
clf = KNeighborsClassifier(n_neighbors=3)
clf.fit(x_train, y_train)  # Training on training data
clf.score(x_test, y_test)  # Testing score on test data
```
