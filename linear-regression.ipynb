```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Making up some data
time_studied = np.array([20, 50, 32, 65, 23, 43, 10, 5, 22, 35, 29, 5, 56]).reshape(-1, 1)
scores = np.array([56, 83, 47, 93, 47, 82, 45, 78, 55, 67, 57, 4, 12]).reshape(-1, 1)
```

```python
# Model training
model = LinearRegression()
model.fit(time_studied, scores)
prediction = model.predict(np.linspace(0, 70, 100).reshape(-1, 1))
```

```python
# Seeing how to do predictions
model.predict(np.array([56, 70]).reshape(-1, 1))
```

```python
# Visualizing results
plt.figure(figsize=(4, 3))
plt.scatter(time_studied, scores)
plt.plot(np.linspace(0, 70, 100), prediction, "red")
plt.show()
```

```python
# Testing model performance
from sklearn.model_selection import train_test_split
```

```python
# Removing outliers
time_studied = np.array([20, 50, 32, 65, 23, 43, 10, 5, 22, 35, 29, 5, 56]).reshape(-1, 1)
scores = np.array([56, 83, 47, 93, 47, 82, 45, 23, 55, 67, 57, 4, 89]).reshape(-1, 1)
```

```python
# Visualizing new data shape
plt.figure(figsize=(4, 3))
plt.scatter(time_studied, scores)
plt.show()
```

```python
# Spliting data
time_train, time_test, score_train, score_test = train_test_split(time_studied, scores, test_size=0.2)
```

```python
# New model
model = LinearRegression()
model.fit(time_train, score_train)  # Only fits model to training data
prediction = model.predict(np.linspace(0, 70, 100).reshape(-1, 1))
model.score(time_test, score_test)  # Tests model accuracy for prediciting new data
```

```python
# Visualizing new results
plt.figure(figsize=(4, 3))
plt.scatter(time_studied, scores)
plt.plot(np.linspace(0, 70, 100), prediction, "red")
plt.show()
```
